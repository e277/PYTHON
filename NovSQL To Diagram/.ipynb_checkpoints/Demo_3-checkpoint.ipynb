```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/Ezra Muir/Documents/Training-Work/Python/Nov_Learn/SQL To Diagram/transaction.csv')
df
```

# Line Charts

```python
plt.figure(figsize=(10,8))
sns.lineplot(x="id",y="cust_id",data=df,linewidth = 5)
plt.title("Transaction Id VS Customer Id")
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=df['id'],linewidth = 1.5 , label = 'Transaction Id')
sns.lineplot(data=df['acc_id'],linewidth = 1.5 , label = 'Transaction Id')
plt.show()
```

# Scatter Plots

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='id',y='acc_id',data=df)
plt.show()

# plt.figure(figsize=(10,10))
# sns.scatterplot(x='id',y='acc_id',data=df, color='#FF0000')
# plt.show()
```

# Bar Plot

```python
plt.figure(figsize=(10,7))
# sns.barplot(df['tran_dt'], df['tran_amt']) # Have warning
sns.barplot(data=df, x='tran_dt', y='tran_amt') 
plt.show()
```

```python
# Horizontal Bar plot
plt.figure(figsize=(7,10))
sns.barplot(data=df, x='tran_amt', y='tran_dt') 
plt.show()
```
